```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
os.chdir(os.path.abspath('..'))

import logging
logging.basicConfig(level=logging.INFO, format='%(levelname)s - %(message)s')

import pandas as pd
from IPython.display import Image, display
from src.data.loader import load_config, resolve_path
from src.models.classification import WeatherClassifier

config = load_config()
df = pd.read_parquet(resolve_path(config['data']['cleaned_parquet']))
print(f'✅ Loaded: {df.shape}')
```

## 1. Huấn luyện & Đánh giá


```python
classifier = WeatherClassifier(config)
results = classifier.run(df)
classifier.save_results(results['results_df'])
```
